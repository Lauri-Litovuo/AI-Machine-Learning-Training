```python
# import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Drop the 'index' column
```

```python
no_index = df.drop(['index'], axis='columns')
no_index
```

```python
# Training and test data split (70:30)
from sklearn.model_selection import train_test_split

X = no_index.loc[:, no_index.columns != 'type']
y = no_index['type']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=10, stratify=y)

print("Training set length:", len(X_train), "\nTesting set length:", len(X_test))
print("Training labels length:", len(y_train), "\nTesting labels length:", len(y_test))
```

```python
# Display class distribution in training and test data

train_cd = y_train.value_counts()
print("Class distribution in training data:")
print(train_cd)

test_cd = y_test.value_counts()
print("Class distribution in test data:")
print(test_cd)
```

```python
# Extracting labels and sizes from train_cd
value_labels = train_cd.index
sizes = train_cd.values

formatted_labels = [f"{label} ({count})" for label, count in zip(value_labels, sizes)]
palette_color = sns.color_palette('muted')
plt.figure(figsize=(10,10))
plt.pie(train_cd, colors=palette_color, labels=formatted_labels, autopct='%.1f%%')
plt.show()
```

```python
# Multinomial logistic regression model
# Training the model
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000, multi_class='multinomial', solver='lbfgs')
#model.fit(X_train, y_train)
```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train the model
model.fit(X_train_scaled, y_train)
```

```python
# Prediction on the training data
from sklearn import metrics
from sklearn.metrics import confusion_matrix

prediction_train = model.predict(X_train_scaled)
accuracy = metrics.accuracy_score(prediction_train, y_train)
print("Accuracy:", accuracy)

cm_train = confusion_matrix(y_train, prediction_train)
print("Confucion matric on training data:")
print(cm_train)
```

```python
# Prediction on the test data
prediction_test = model.predict(X_test_scaled)
accuracy = metrics.accuracy_score(prediction_test, y_test)
print("Accuracy:", accuracy)

cm_test = confusion_matrix(y_test, prediction_test)
print("Confucion matric on training data:")
print(cm_test)
```

```python
# Calculate the confusion matrix

cm_hm = confusion_matrix(y_test, prediction_test)

# Create a heatmap of the confusion matrix
plt.figure(figsize=(8,8))
sns.heatmap(cm_test,
            annot=True,
            fmt='g',
            cmap='Blues',
            cbar=False
            )
plt.xlabel('Prediction', fontsize=13)
plt.ylabel('Actual labels', fontsize=13)
plt.title('Confusion Matrix', fontsize=15)
```
