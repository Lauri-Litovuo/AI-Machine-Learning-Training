```python
#creating a dict 'company price' where key = car's company and value = median of their price

...

# company_price
```

```python
median_prices = CarData.groupby('CarCompany')['price'].median()
# print(median_prices)
company_price = median_prices.to_dict()
company_price

```

```python
# division in the buckets of low i.e. below 10000, medium i.e. range b/w 10000 and 20000 and high i.e. above 20000

# price_ranges = {'low': (0,10000), 'medium': (10001, 20000), 'high': (20001, 50000)}

company_price_range = {}

for company, price in company_price.items():
  if price <= 10000:
    company_price_range[company] = 'low'
  elif price > 10000 and price <= 20000:
    company_price_range[company] = 'medium'
  else:
    company_price_range[company] = 'high'

print(company_price_range)

# mapping the company_price on the 'CarCompany' column in the dataset

CarData['CarCompany'] = CarData['CarCompany'].map(company_price_range)
CarData = CarData.drop('CarName', axis=1)
CarData.head()

# checking the column's unique values

#CarData.CarCompany.unique()
# CarData.company_price.unique()
```

```python
#creating dummy variable for catwegorical variables


# categorical_vars = CarData.select_dtypes(include=['object']).columns
CarData = pd.get_dummies(CarData)

#checking the dataset
CarData.head()
```

```python
# importing required library
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively

CarData_train, CarData_test = train_test_split(CarData, test_size=0.3, random_state=100)

# shape of the train and test sets
print(CarData_train.shape)
print(CarData_test.shape)
```

```python
#importing required function for scaling

from sklearn.preprocessing import MinMaxScaler

scale = MinMaxScaler()
```

```python
y_train = CarData_train['price']
X_train = CarData_train.drop('price', axis=1)

y_test = CarData_test['price']
X_test = CarData_test.drop('price', axis=1)

X_train_scaled = scale.fit_transform(X_train)
X_test_scaled = scale.transform(X_test)
```

```python
# Train your model
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train_scaled, y_train)
```

```python
# Making predictions
predicted_price = model.predict(X_test_scaled)
```

```python
# Show predictions
predicted_price
```

```python
# Plotting y_test and y_pred to understand the spread.
sns.scatterplot(x=y_test, y=predicted_price)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, predicted_price)
print(r2)
```
